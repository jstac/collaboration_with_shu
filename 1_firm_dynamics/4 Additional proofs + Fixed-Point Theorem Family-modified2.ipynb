# 1 One proposition and its proof

From the lecture about the Finite Markov Chain, we know the definiton of the stochastic matrix:

## Def. Stochastic Matrix*

Let $\mathbb{1}$ be a $m$-dimenstional identity vector, that is, 
$$
\mathbb{1}=
\begin{pmatrix}
1 \\
1 \\
\vdots\\
1
\end{pmatrix} \tag{1}
$$
and $\mathbb{0}$ be a $m$-dimensional zero matrix, that is, 
$$
\mathbb{0}=
\begin{pmatrix}
0 & 0 & \cdots & 0 \\
0 & 0 & \cdots & 0 \\
\vdots\\
0 & 0 & \cdots & 0
\end{pmatrix} \tag{2}
$$
A square matrix $P=(p_{i,j})_{m \times m}$ is a stochastic matrix if $P$ satifies two conditions:

1. $P \geq \mathbb{0}$ and
2. $P \cdot \mathbb{1}=\mathbb{1}$.



We have the following proposition:

## Proposition 1

If $P$ is a stochastic matrix, then the $k$-th power $P^k$ for all $k \in N$ is also a stochastic matrix. 

## Proof by induction*.

Let $\mathbb{0}$ be a $m$-dimensional square matrix and $P$ be a $m$-dimensional stochastic matrix.

#### proof.when_n=1

When $n= 1$, we have$P^1=P$.

Since $P$ is a stochastic matrix, $P^1$ is also a stochastic matrix, which satisfies two conditions:

1. $P^1 \geq \mathbb{0}$ and 
2. $P^1 \cdot \mathbb{1}=\mathbb{1}$.

#### Proof.when_n=k+1

Assume that the matrix $P^k$ is also a stochastic matrix, that is,

1. $P^k \geq \mathbb{0}$ and 
2. $P^k \cdot \mathbb{1}=\mathbb{1}$.

For $P^{k+1}= P^1 \times P^k$, since we have $P^1 \geq \mathbb{0}$ and $ P^k \geq \mathbb{0}$, by definition of matrix muliplication operation, we can conclude
$$
P^{k+1} \geq \mathbb{0} \tag{3}
$$
And because we have $P^k \cdot \mathbb{1}=\mathbb{1}$ and $P \cdot \mathbb{1}=\mathbb{1}$, we can obtain
$$
P^{k+1} \mathbb{1} =P \times P^k \mathbb{1} \\
= P \cdot\mathbb{1} \\
=\mathbb{1} \tag{4}
$$
By the definition of a stochastic matrix, we can conclude that the matrix $P^{k+1}$ is also a stochastic matrix.

By induction, we can conclude that if $P$ is a stochastic matrix, then the $k$-th power $P^k$ for all $k \in N$ is also a stochastic matrix. 



# 2 Brouwer's and Kakutani's Fixed-point theorem

## Def. (Fixed Point; Function)

Given a set $X$ and a function $f: X \to X$, $x^* \in X$ is a fixed point of $f$ iff $ f(x^*)=x^* $.

## Def. (Fixed Point; Correspondence)

Given a set $X$ and a correspondence $F: X \implies X$, $x^* \in X$ is a fixed point of $F$ iff $ F(x^*)=x^* $.

## Proposition 2 (Brouwer's Fixed-point theorem)

If $X$ is a nonempty, compact, and convex subset of $ \mathbb R^k$ for some integer $k$, and $f: X \to X$ is continuous, then there exists a fixed point of $f$.

## Proposition 3 (Kakutani's Fixed-Point Theorem)

If $X$ is a nonempty compact, and convex subset of $\mathbb R^k$ for some integer $k$, and $F: X \implies X $ (Not a function) is nonempty and convex valued and upper semi-continuous, then there exists a fixed point of $F$.

### Remark

These two are equivalent, that is, Brouwer's FP theorem is implied by Kakutani's, and Kakutani's can be proved if one assumes Brouwer's.

### Intuitive demonstration of Brouwer's FP Theorem

Suppose $X$ is a circle (interior and boundary) in $ \mathbb R^2$, and $f: X \to X$ has no fixed point.

For each point $x \in X$, find $f(x)$ and draw a ray from $f(x)$ through $x$, label the point on the boundary hit by this ray as $g(x)$.

As long as $f(x) \neq x$, this is a well-defined process for defining $g(x)$.

If $x$ moves a bit, then $f(x)$ moves only a bit. Hence, the ray moves only a bit and the intersection point moves only a bit. So intuitively, we can conclude that $g$ is continous.

Moreover, if $x$ lies on the boundary of $X$, then $g(x) = x$.

But then $g$ is a continous map from $X$ to its boundary that is the identity on the boundary. And intuitively one cannot do this. You must "tear" $X$ somewhere to map the whole circle onto its boundary in a way that leaves the boundary fixed.

The existence of a fixed point rule this thing(? what exactly) out because:

If $f(x)=x$, then it isn't clear how to define $g(x)$.

More to the point, slight movements in $x$ around such a fixed point can mean a big movement in $g(x)$; the "proof" that $g$ is continous depends on some separation between $x$ and $f(x)$.

#### Remark

##### 1. This is not a proof, but we can turn it into one:

A result in Borsuk's Lemma (*?*) shows that this sort of continous map is impossible (in $ \mathbb R^2$ and in higher dimensions)

##### 2. More direct proofs are possible, and the most accessible are "constructive" or computational:

Using Sperner's Lemma (*?*), we can show how to find "approximate fixed points" to any degree of approximation desired, and then pass to a limit. (See Border (1990) or Scarf (1973)). We can prove Brouwer in this fashion, and use the Green's Theorem (*?*) as the basis of a proof.

#### Corollary 1

# 3 Existence of stationary distribution of a stochastic matrix

## Lemma 1 (Limits preserve weak inequalities)

Let $\{ x_n \}$ be a convergent sequence with the limit $\lim_{n \to \infty} x_n = x$, and $a \in \mathbb R$.

If $x_n \leq a $ holds for any $n \in \mathbb N$, then we have $x \leq a$.

## Lemma 2 (Sufficient and Necessary Condition for a closed set)

A set $C \subset \mathbb R^m$ is closed iff for every convergent sequence $(z_n)$ that entirely lie in $C$ s.t. its limit is also in $C$.

## Def. (Stationary distribution)

A distribution $\psi^*$ on $S$ is called **stationary** for $P$ if 
$$
\psi^* = \psi^* P \tag{5}
$$

## Proposition 4. (Stochastic matrix vs. Existence of Stationary distribution)

Assume that the state space $S$ is finite (if not more assumptions are required).

Every stochastic matrix $P$ has at least one stationary distribution.

## Proposition 5. (Equivalence of the Prop.4)

Without loss of generality, let $\mathbb{0}$ be the $m$-dimensional zero vector,  $S=\{w_1, w_2,...,w_m\}$ be the state space, $\Psi= \{ \psi =(x_1, x_2, ..., x_m) \in \mathbb R^m| \psi \geq \mathbb{0} \ and \ \langle \psi, \mathbb{1} \rangle =1   \}$ be the set of all possible distributions, that is, $\psi_t \in \Psi $ for any t,  and $P$ be a $m$-dimension stochastic matrix.

Suppose that $P: \psi \to \psi P$. Since $\psi=\Psi$ and $\psi P \subseteq \Psi$, w.l.g., we can write $P:\Psi \to \Psi$ such that $P(\psi_t)= \psi_{t+1}$,

then there exists a fixed point of $P$.

### Analysis

If we can prove $\Psi$ is a nonempty, compact, and convex subset of the Euclidean space $R^m$ and $P: \Psi \to \Psi$ such that $P(\psi_t)= \psi_{t+1}$ is continuous, then by Brouwer's Fixed-point theorem, we prove the proposition, and then the existance of a stationary distribution for any stochastic matrix.

### Proof.

#### proof.nonempty_set

Since at least $(0,..,0,1) \in \Psi$, so $\Psi$ is not an empty set.

#### proof.bounded_set

From the set $\Psi$, since $x_i \geq 0$ and $\langle \psi, \mathbb{1} \rangle =1$, we have $0\leq x_i \leq 1$.

For any $\psi \in \Psi$, we have
$$
\mathbb{0} \leq \psi \leq \mathbb{1} \tag{6}
$$
By definition of a bounded set, we can conclude that $\Psi$ is bounded.

#### proof.closed_set*

For any convergent sequence $(\xi_n) \in \Psi$ with its limit $\lim_{n \to \infty} \xi_n = \xi \in \mathbb R^m  $.

##### proof.$\xi \geq \mathbb{0}$

By definition of the set $\Psi$, we know that 
$$
\xi_n \geq \mathbb{0} \ for \ all \ n \tag{7}
$$
By Lemma 1, we have
$$
\xi \geq 0 \tag{8}
$$

##### Proof.$\langle \xi, \mathbb{1} \rangle =1$

Let $G$ be a function defined on $\mathbb R^m$, that is, $G: \mathbb R^m \to \mathbb R$ s.t. $G(z)= \langle z , \mathbb{1}  \rangle $, where $z \in \mathbb R^m$.

Obviously, $G$ is continuous on $\mathbb R^m$.

By the definition of the set $\Psi$, we know that
$$
G(\xi_n) = \langle \xi_n , \mathbb{1}  \rangle = 1 \ holds \ for \ all \ n \tag{9}
$$
 Thus, we have
$$
\lim_{n\to \infty} G(\xi_n) = \lim_{n\to \infty} \langle \xi_n , \mathbb{1}  \rangle = 1 \tag{10}
$$
Since $G$ is continuous on $\mathbb R^m$, we can put the limit into the function without change the final outcome
$$
G(\lim_{n\to \infty} \xi_n) =  \langle \lim_{n\to \infty} \xi_n , \mathbb{1}  \rangle = \langle \xi , \mathbb{1}  \rangle =1 \tag{11}
$$
By the definition of the set $\Psi$, we have
$$
\xi \in \Psi \tag{12}
$$
By lemma 2, we can conclude that the set $\Psi$ is closed in the Euclidean space $\mathbb R^m$.

#### proof.compact_set

Given $\Psi$ is bounded and closed in the Euclidean space $\mathbb R^m$, we know that $\Psi$ is compact.

#### proof.convex_set*

For any two vectors $\psi, \psi' \in \Psi$ and $\alpha \in [0,1]$, we have
$$
\alpha \psi + (1-\alpha)\psi' \tag{13}
$$
By the definition of the set $\Psi$, we have
$$
\psi \geq \mathbb{0} \ and \ \psi' \geq \mathbb{0} \tag{14}
$$
By definition of vector's scalar multiplication and addition operations, we obtain
$$
\alpha \psi + (1-\alpha) \psi' \geq 0 \ where \ \alpha \in [0,1] \tag{15}
$$
By the definition of the set $\Psi$, we have
$$
\langle \psi, \mathbb{1} \rangle =1 \ and \ \langle \psi', \mathbb{1} \rangle =1 \tag{16}
$$
By the definition of real number's multiplication and addition operations, we have
$$
\langle \alpha \psi + (1-\alpha) \psi', \mathbb{1} \rangle = \alpha \langle \psi, \mathbb{1} \rangle + (1-\alpha) \langle \psi', \mathbb{1} \rangle \\
= \alpha \cdot 1 + (1-\alpha) \cdot 1 \\
= \alpha + 1 -\alpha \\
= 1 \tag{17}
$$
According to the definition of the set $\Psi$, we can conclude that 
$$
[ \alpha \psi + (1-\alpha)\psi' ] \in \Psi \tag{18}
$$
Based on the definition of the convex set, we can conclude that the set $\Psi$ is convex.

#### proof.continous_function

For any vector $\psi_t = (x^t_1, x^t_2, ..., x^t_m) \in \Psi$, we have
$$
P(\psi_t) = P(x^t_1, x^t_2, ..., x^t_m)\\
= (p_1(x^t_1, x^t_2, ..., x^t_m), p_2 (x^t_1, x^t_2, ..., x^t_m),..., p_m(x^t_1, x^t_2, ..., x^t_m))           \\
=(\sum^m_{i=1}x^t_i p_{i,1}, \sum^m_{i=1}x^t_i p_{i,2},..., \sum^m_{i=1}x^t_i p_{i,m})=(x^{t+1}_1, x^{t+1}_2, ..., x^{t+1}_m)=\psi_{t+1} \tag{19}
$$
where $p_j (x^t_1, x^t_2, ..., x^t_m)=\sum^m_{i=1}x^t_i p_{i,m}=x^{t+1}_j$.

Given previous conditions, we know the function $p_j(\psi_t)=x^{t+1}_j$ is continuous on $\mathbb R^m_+$, and hence, $P(\psi_t)=\psi_{t+1}$ should also be continuous on $\Psi$.

#### proof.fixed_point

By Brouwers's Fixed-Point theorem, we can conclude that there exists some $\psi^* \in \Psi$ such that
$$
P(\psi^*)= \psi^* \tag{20}
$$
Which is what we want to prove for the proposition.

Since the proposition is equivalent to the existence theorem, we can also conclude that every stochastic matrix $P$ has at least one stationary distribution.

## A1 The simplest Fixed-point theorem

### Proposition 6

Suppose $f:[0,1] \to [0,1]$ is continuous. Then there exists a fixed point of $f$.

#### Proof

Define the function $\phi: [0,1]\to \mathbb R$ by $\phi(x) = f(x)- x$.

The function $\phi$ is continous, and $\phi(0) = f(0)-0= f(0) \geq 0$ while $\phi(1)=f(1)-1 \leq 0$.

By Intermediate-value theorem, there exists some point $x^* \in [0,1]$ s.t.
$$
\phi(x^*) = 0 \tag{21}
$$
Since $\phi(x) = f(x)- x$, we have
$$
f(x^*)- x^*=0 \ or \ f(x^*)= x^* \tag{22}
$$

#### Remark

This is the simplest of all fixed-point theorems.

## A2 General Forms of the Fixed-point theorem

More generally, fixed-point theorems have two basic forms.

### Proposition 7 (FP form for function)

If $f$ is a function with domain $X$ and range $X$ [and then conditions on $X$ and $f$ are given], then there exists a fixed point of $f$.

### Proposition 8 (FP form for correspondence)

If $F$ is a correspondence with domain $X$ and range $X$ (that is, $F(x) \subseteq X$ for all $x \in X$), [and then conditions on $X$ and $F$ are given], then there exists a fixed point of $F$.

#### Remarks

The fixed-point theorem will sometimes say about the structure of the set of fixed points.

## A5 Banach's and Nadler's Fixed-Point Theorems

Because Banach's is sometimes used in context where the domain (and range) of the function are more complex than finite-dimensional Euclidean space, we will give it in greater generality.

### Def. (Contraction mapping)

The function $f$ is called a **contraction mapping** if, for some $\alpha <1$, $f: X \to X$ satisfies:
$$
d(f(x), f(x')) \leq \alpha d(x, x') \ for \ all \ x,x' \in X \tag{23}
$$

### Proposition 11 (Banach's; Function)

If $f: X \to X$ is contraction mapping defined on a complete metric space $(X,d)$, then .there exists an unique fixed point of $f$.

#### Proof*.

##### Existence

Take any point from $X$ and label it $x^0$.

Le $x^1 = f(x^0)$ and, inductively, $x^n = f(x^n)$.

Let $A= d(x^0, x^1)$, so that, inductively, $d(x^n, x^{n+1}) \leq \alpha^n A$.

Applying the triangle inequality yields
$$
d(x^n, x^m) \leq A( \alpha^n + \cdots + \alpha^{m-1}) \leq \frac{\alpha^nA}{1- \alpha} \tag{24}
$$
where $m > n$.

By the definition of the Cauchy sequence, the sequence $\{ x^n \}$ is a Cauchy sequence.

Since the space $X$ is complete, this sequence has a limit $x^*$, that is,
$$
\lim_{n \to \infty} d(x^n, x^*) = 0 \tag{25}
$$
Since $d(f(x^n), f(x^*)) \leq \alpha d(x^n, x^*)$, we know that
$$
\lim_{n \to \infty} d(f(x^n), f(x^*)) = 0 \tag{26}
$$
which is
$$
\lim_{n \to \infty} d(x^{n+1}, f(x^*)) = 0 \tag{27}
$$
so we have
$$
\lim_{n \to \infty} x^n = f(x^*) \tag{28}
$$

##### Uniqueness

Using the triangle inequality, one can show that in a metric space, a convergent sequence can have only one limit, so $f(x^*)=x^*$.

Suppose $x^*_1$ and $x^*_2$ are two fixed points of $f$, that is, 
$$
f(x^*_1)=x^*_1 \ and \ f(x^*_2)=x^*_2 \tag{29}
$$
But since $f$ is a contraction mapping, by definition of contraction mapping, we have
$$
d(x^*_1, x^*_2) =d(f(x^*_1), f(x^*_2)) \leq \alpha d(x^*_1, x^*_2) \tag{30}
$$
for $\alpha < 1$.

This implies that $d(x^*_1, x^*_2) =$ or equivalently $x^*_1 = x^*_2$, so the fixed point is unique

### Def. (Contractive correspondence)

The correspondence $F$ is **contractive** if, for some $\alpha <1$, $F: X \to Z$ satisfies $d_H(F(x), F(x')) \leq \alpha d(x,x')$ for all $x, x^0 \in X$.

### Proposition 12 (Nadler's; Correspondence)

If $F$ is a contractive, nonempty-valued, and compact-valued correspondence defined on a complete metric space, then there exists a fixed point of $F$.

## A6 Tarski's Fixed-Point theorem

Tarski's Fixed-Point theorem concerns functions on ordered sets. So we should introduce the definition of ordered sets firstly.

### Def. (Partially ordered set/ poset)

A partially ordered set, or poset, consists of two things:

1. An abstract set $X$ and 
2. A binary relation $\succsim$ on $X$ that is reflexive, transitive and anti-symmetric (if $x \succsim y$ and $y \succsim x$, then $x=y$).

### Def. (Upper bound; Supremum)

If $Y$ is a subset of $X$ for some partially ordered set $(X, \succsim)$, an **upper bound** for $Y$ is an element $x \in X$ such that $x \succsim y$ for all $y \in Y$.

A **supremum** for $Y$ is an upper bound $x^*$ of $Y$ that, moreover, satisfies $x \succsim x^*$ for all other upper bounds $x$ of $Y$.

### Def. (Join semi-lattice)

A **join semi-lattice** is a partially ordered set $(X, \succsim)$ such that if $x$ and $x'$ are any two elements of $X$, then the set $\{ x, x' \}$ has a supremum in $X$, which is denoted as $x \vee x'$ and called the **join** of $x$ and $x'$.

### Proposition 13 (Tarski's Fixed-Point Theorem)

Suppose $(X, \succsim)$ is a join semi-lattice with the following additional properties:

1. The set $X$ has a maximal and a minimal element: there exist $\bar{x} \in X$ and $\underline{x} \in X$ such that $\underline{x} \succsim x \succsim \bar{x}$ for all $x \in X$.
2. Every nonempty subset $Z$ of $X$ has a supremum in $X$.

If $f: X \to X$ is a nondecreasing function, meaning $x \succsim x'$ implies $f(x) \succsim f(x')$, then there exists a fixed point of $f$.

Moreover, the supremum of the set of fixed points of $f$ is itself a fixed point of $f$.

### Proposition 14 (Zhou's Fixed-Point theorem)

Suppose $(X, \succsim)$ is a join semi-lattice with the two additional properties (same as Prop. 13): 

1. The set $X$ has a maximal and a minimal element: there exist $\bar{x} \in X$ and $\underline{x} \in X$ such that $\underline{x} \succsim x \succsim \bar{x}$ for all $x \in X$.
2. Every nonempty subset $Z$ of $X$ has a supremum in $X$.

Suppose that $F: X \implies X$ is a nonempty-valued correspondence on $X$, such that,

1. for each $x \in X$, $F(x)$ contains its own supremum and,
2. if $x \succsim  y$, $x' \in F(x)$, and $y' \in F(y)$, then $x' \vee y' \in F(x)$.

Then there exists a fixed point of $F$.
